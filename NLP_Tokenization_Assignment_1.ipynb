```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('owm-1.4')

from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer
```

#Sample Text

```python
text = "NLTK is amazing! I'm learning tokenization, stemming, and lemmatization :) #NLP"
print(text)
```

#Tokenization


```python
from nltk.tokenize import WhitespaceTokenizer

wt = WhitespaceTokenizer()
whitespace_tokens = wt.tokenize(text)

print("Whitespace Tokenization:")
print(whitespace_tokens)

```

#Punctuation-based Tokenization

```python
from nltk.tokenize import wordpunct_tokenize  #separate all punctuations (. , ! "" etc)

punct_tokens = wordpunct_tokenize(text)

print("Punctuation-based Tokenization:")
print(punct_tokens)
```

#Tree Bank Tokenization


```python
from nltk.tokenize import TreebankWordTokenizer  #Tree bank separate all except '.'

treebank = TreebankWordTokenizer()
treebank_tokens = treebank.tokenize(text)

print("Treebank Tokenization:")
print(treebank_tokens)
```

#Tweet Tokenizer


```python
from nltk.tokenize import TweetTokenizer  #similar are word Tokenizer

tweet_tokenizer = TweetTokenizer()
tweet_tokens = tweet_tokenizer.tokenize(text)

print("Tweet Tokenization:")
print(tweet_tokens)
```

#Multi Word Expression (MWE) Tokenization

```python
from nltk.tokenize import MWETokenizer  #separates Multi Word

mwe = MWETokenizer([('machine', 'learning'),
                    ('natural', 'language', 'processing')])

sentence = "I love machine learning and natural language processing"
mwe_tokens = mwe.tokenize(sentence.split())

print("MWE Tokenization:")
print(mwe_tokens)
```

#Porter Stemmer

```python
from nltk.stem import PorterStemmer #convert the words in base word

ps = PorterStemmer()
porter_stems = [ps.stem(word) for word in treebank_tokens]

print("Porter Stemmer:")
print(porter_stems)
```

#Snowball Stemmer

```python
from nltk.stem import SnowballStemmer

ss = SnowballStemmer("english")
snowball_stems = [ss.stem(word) for word in treebank_tokens]

print("Snowball Stemmer:")
print(snowball_stems)
```

#WordNet Lemmatization

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in treebank_tokens]

print("Lemmatization:")
print(lemmatized_words)
```
